```python
class Customer:
    def __init__(self, name, email):
        self.name = name.strip().title()  # Clean and capitalize name
        self.email = email.strip().lower()  # Clean and lowercase email
        self.balance = 0

    def deposit(self, amount):
        self.balance += amount
        print(f"‚Çπ{amount} deposited successfully. Current Balance: ‚Çπ{self.balance}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"‚Çπ{amount} withdrawn successfully. Current Balance: ‚Çπ{self.balance}")
        else:
            print("Insufficient balance!")

    def check_balance(self):
        print(f"{self.name}'s Current Balance: ‚Çπ{self.balance}")


class Bank:
    def __init__(self):
        self.customers = {}  # key: (name, email), value: Customer object

    def add_customer(self, name, email):
        key = (name.strip().lower(), email.strip().lower())
        if key in self.customers:
            print("Customer already exists!")
        else:
            self.customers[key] = Customer(name, email)
            print("New customer added successfully!")

    def get_customer(self, name, email):
        key = (name.strip().lower(), email.strip().lower())
        return self.customers.get(key, None)


# -------- MAIN PROGRAM --------
bank = Bank()

while True:
    print("\nüè¶ BANK MENU")
    print("1. Add New Customer")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Check Balance")
    print("5. Exit")

    choice = input("Enter your choice: ").strip()

    if choice == '1':
        name = input("Enter name: ").strip()
        email = input("Enter email: ").strip()
        bank.add_customer(name, email)

    elif choice == '2':
        name = input("Enter name: ").strip()
        email = input("Enter email: ").strip()
        customer = bank.get_customer(name, email)
        if customer:
            amount = int(input("Enter deposit amount: "))
            customer.deposit(amount)
        else:
            print("Customer not found!")

    elif choice == '3':
        name = input("Enter name: ").strip()
        email = input("Enter email: ").strip()
        customer = bank.get_customer(name, email)
        if customer:
            amount = int(input("Enter withdrawal amount: "))
            customer.withdraw(amount)
        else:
            print("Customer not found!")

    elif choice == '4':
        name = input("Enter name: ").strip()
        email = input("Enter email: ").strip()
        customer = bank.get_customer(name, email)
        if customer:
            customer.check_balance()
        else:
            print("Customer not found!")

    elif choice == '5':
        print("Thank you for using the bank system!")
        break

    else:
        print("Invalid choice! Please try again.")
```
